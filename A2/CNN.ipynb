```python
import torch
import torch.nn as nn
import torch.backends.cudnn as cudnn
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
use_cuda = torch.cuda.is_available()
```

```python
from dataset import dataset
from AlexNet import AlexNet
from train_test import start_train_test
from train_test import test
```

```python
trainloader, testloader, outputs, inputs = dataset('dog-breed')
print ('Output classes: {}\nInput channels: {}'.format(outputs, inputs))
net = AlexNet(num_classes = outputs,inputs=inputs)
file_name = 'alexnet-'
if use_cuda:
    net.cuda()
    net = torch.nn.DataParallel(net, device_ids=range(torch.cuda.device_count()))
    cudnn.benchmark = True

criterion = nn.CrossEntropyLoss()
train_loss, test_loss = start_train_test(net, trainloader, testloader, criterion)
```

```python
from transform import transform_testing
import torchvision
import pandas as pd
import csv

testset = torchvision.datasets.ImageFolder('./data/test/', transform=transform_testing())
testloader = torch.utils.data.DataLoader(testset, batch_size=256, shuffle=False, num_workers=5)

df = pd.read_csv('./sample_submission.csv',nrows=0)
image_names = os.listdir('./data/test/images/')
output_list = []
name_counter = 0

df.to_csv('./submission.csv',sep=',',index=False)
net.eval()
for data in testloader:   
    inputs, labels = data
    outputs = net(inputs)
    soft_max = torch.nn.Softmax()
    prob = soft_max(outputs.data)
    for index in range(len(prob)):
        temp = prob[index].tolist()
        temp.insert(0,image_names[name_counter])
        name_counter +=1
        output_list.append(temp)
with open("./submission.csv",'a') as resultFile:
    wr = csv.writer(resultFile)
    wr.writerows(output_list)
            
    

```
